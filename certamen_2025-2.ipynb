### Contexto: Problema MERP-IoT (Multi-hop Energy-efficiency Routing Problem in IoT)

Una empresa desea desplegar una red IoT inalámbrica para monitoreo ambiental en una ciudad. La red está compuesta por:

* Un conjunto de **nodos sensores** de bajo consumo energético distribuidos en el territorio.
* Uno o varios **nodos sumidero (sink / gateway)** que recolectan los datos y los envían a la nube.
* Enlaces inalámbricos entre nodos, que sólo existen si la potencia de transmisión permite una comunicación confiable dentro de un cierto radio.

Cada nodo tiene:

* Una **batería limitada** (energía residual inicial conocida).
* Capacidad para **reenviar paquetes** de otros nodos (enrutamiento multi-salto / multi-hop).
* Restricciones de **tasa máxima de transmisión** y **cola de paquetes**.

Cada enlace ((i,j)) tiene asociadas:

* Un **costo energético** por transmitir un paquete en ese enlace.
* Un **retardo (delay)** estimado.
* Una **probabilidad de entrega de paquete (PDR)** que refleja la calidad del enlace.

El operador de la red desea diseñar **rutas multi-hop** desde cada sensor hasta un sink. Para evaluar la calidad de una solución de enrutamiento, se consideran las siguientes **variables de desempeño**:

1. **Packet Delivery Ratio (PDR):** proporción de paquetes que llegan correctamente al destino respecto del total enviado.
2. **Average Energy:** **energía promedio** consumida por los nodos de la red.
3. **Average Latency:** **latencia promedio** de los paquetes (tiempo medio de viaje desde el nodo origen al sink).
4. **Total Energy:** **energía total** consumida por todos los nodos durante la transmisión de los flujos.

El objetivo es **encontrar configuraciones de rutas** que logren un **buen compromiso** entre estas métricas: mantener un **PDR alto**, minimizar la **energía total** y la **energía promedio por nodo**, y al mismo tiempo reducir la **latencia promedio**.

 el espacio de soluciones posibles (conjunto de rutas posibles para todos los nodos) crece de forma explosiva con el número de nodos. Por ello, la empresa está interesada en utilizar un **método de computación evolutiva** que permita aproximar buenas soluciones en tiempos razonables.

```python
import pandas as pd
import numpy as np

# Configuración general
AREA_SIZE = 50   # Área: 50x50 unidades
INSTANCIAS = [20, 30, 50]  # Cantidad de nodos por instancia

# Semilla base para reproducibilidad
np.random.seed(17) # el numero de la semilla debe ser los ultimos 2 digitos de su rol

for N in INSTANCIAS:
    x = np.random.uniform(0, AREA_SIZE, N)
    y = np.random.uniform(0, AREA_SIZE, N)

    nodes_df = pd.DataFrame({
        'id': range(N),
        'x': x,
        'y': y
    })

    # Nombre de archivo según cantidad de nodos
    filename = f"nodes_{N}.csv"

    nodes_df.to_csv(filename, index=False, header=False)

    print(f"Archivo {filename} generado con {N} nodos en un área de {AREA_SIZE}x{AREA_SIZE}.")
```
